```python
# Creació de les caselles representades amb un diccionari on les claus seran les coordeandes sobre el plà
# i els valors un diccionari de característiques. N'inicialitzem els valors a zero.

num_rows = 3
num_cols = 4

squares = {(row, col): {'value': 0.0, 'update': 0.0} for row in range(1, num_rows + 1) for col in range(1, num_cols + 1)}

# Assignem valors manualment al bloc i a les caselles de final del joc.
squares[(2, 2)]['value'] = '#'
squares[(3, 4)]['value'] = 1
squares[(2, 4)]['value'] = -1
```

```python
# Definició de les possibles direccions
(north, south, east, west) = ((1, 0), (-1, 0), (0, 1), (0, -1))

# Política de cada cel·la (aquesta política és òptima per recompenses entre 0 i -0.05)
# Per recompenses positives o recompenses molt negatives s'han de fer reassignacions
squares[(1, 1)]['policy'] = north
squares[(1, 2)]['policy'] = west
squares[(1, 3)]['policy'] = west
squares[(1, 4)]['policy'] = west
squares[(2, 1)]['policy'] = north
squares[(3, 1)]['policy'] = east
squares[(3, 2)]['policy'] = east
squares[(3, 3)]['policy'] = east
squares[(2, 3)]['policy'] = north
```

```python
# Definim els factors (En aquest exmple es pot canivar la recompensa amb valors de 0 a 0.05 aprox
# perquè la política segueixi essent òptima). El soroll, el descompte i el hit es poden canviar sempre

reward = -0.04
hit = .8
noise = .2
discount = 1
wall = '#'
```

```python
# Definim el soroll
for attrs in squares.values():
    if attrs.has_key('policy') and (attrs['policy'] == north or attrs['policy'] == south):
        attrs['noise'] = east, west
    elif attrs.has_key('policy'):
        attrs['noise'] = north, south
```

```python
def bellman():
    """ Iteracions de bellman que actualitzen els valors fins a convergir """
    
    func = lambda x, y: x + y
    
    for square, attrs in squares.iteritems():
        
        # Només apliquem bellman a les caselles que tenen política (no blocs i no final)
        if attrs.has_key('policy'):
            
            # Calculo la cel·la de transició
            transition = squares[tuple(map(func, attrs['policy'], square))]
            
            # Cel·les de soroll
            noise_1 = tuple(map(func, attrs['noise'][0], square))
            noise_2 = tuple(map(func, attrs['noise'][1], square))
            
            # Valors de soroll per defecte
            noise_1_value = attrs['value']
            noise_2_value = attrs['value']
            
            # Actualització del soroll si escau
            if squares.has_key(noise_1) and squares[noise_1]['value'] != wall:
                noise_1_value = squares[noise_1]['value']
                
            if squares.has_key(noise_2) and squares[noise_2]['value'] != wall:
                noise_1_value = squares[noise_2]['value']

            # Actualitzem el valor de la casella per la propera iteració
            attrs['update'] = hit * (reward + (discount * transition['value']))
            
            # Afegim el soroll al valor
            attrs['update'] += ((noise / 2) * (reward + (discount * noise_1_value)))
            attrs['update'] += ((noise / 2) * (reward + (discount * noise_2_value)))
            
    # Actualitzem els valors per la següent iteració
    for attrs in squares.values():
        if attrs.has_key('policy'):
            attrs['value'] = attrs['update']
                                        
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx

def show_result(num_iterations):
    """ Calcula els valors bellman i mostra els resultats gràficament """
    
    # Funció interna
    def func(argtype):
        return lambda x: type(graph.node[x]['value']) == argtype
    
    for i in range(num_iterations):
        
        # Càlculs
        bellman()

        # Graph
        graph = nx.DiGraph()
        graph.add_nodes_from(squares.items())

        # Figura
        fig, ax = plt.subplots(figsize=(9.8, 7.6))
        fig.suptitle(u"\nIteració {0}".format(i + 1), fontsize=20)
        ax.xaxis.set_visible(False) 
        ax.yaxis.set_visible(False)

        # Posicions dels nodes a la gràfica
        graph.node[(1, 1)]['pos'] = 0.0, 0.0
        graph.node[(1, 2)]['pos'] = 1.0, 0.0
        graph.node[(1, 3)]['pos'] = 2.0, 0.0
        graph.node[(1, 4)]['pos'] = 3.0, 0.0
        graph.node[(2, 1)]['pos'] = 0.0, 1.0
        graph.node[(2, 2)]['pos'] = 1.0, 1.0
        graph.node[(2, 3)]['pos'] = 2.0, 1.0
        graph.node[(2, 4)]['pos'] = 3.0, 1.0
        graph.node[(3, 1)]['pos'] = 0.0, 2.0
        graph.node[(3, 2)]['pos'] = 1.0, 2.0
        graph.node[(3, 3)]['pos'] = 2.0, 2.0
        graph.node[(3, 4)]['pos'] = 3.0, 2.0
        
        # Afegint les posicions
        pos = nx.get_node_attributes(graph, 'pos')

        # Política
        nodes = filter(lambda x: graph.node[x].has_key('policy'), graph.nodes())
        edges = map(lambda x: (x, tuple(map(lambda a, b: a + b, x, graph.node[x]['policy']))), nodes)
        graph.add_edges_from(edges)

        # Nodes
        nodes_default = filter(func(float), graph.nodes())
        nodes_goal = filter(func(int), graph.nodes())
        nodes_wall = filter(func(str), graph.nodes())
        
        # Llegendes
        labels_default = {x: round(graph.node[x]['value'], 3) for x in nodes_default}
        labels_goal = {x: graph.node[x]['value'] for x in nodes_goal}
        labels_wall = {x: graph.node[x]['value'] for x in nodes_wall}
        
        # Opcions
        nx.draw_networkx(graph, pos, nodelist=nodes_default, labels=labels_default, with_labels=1, node_size=3000, node_color='gray', node_shape='s')        
        nx.draw_networkx(graph, pos, nodelist=nodes_goal, labels=labels_goal, with_labels=1, node_size=4200, node_color='yellow', node_shape='h')
        nx.draw_networkx(graph, pos, nodelist=nodes_wall, labels=labels_wall, with_labels=1, node_size=6000, node_color='black', node_shape='s')

        plt.show()
        
    # Netejem els valors per poder executar amb diferent nombre d'iteracions
    for attrs in squares.values():
        if attrs.has_key('policy'):
            attrs['value'], attrs['update'] = 0.0, 0.0
```

```python
# El num d'iteracions es pot canviar
# Amb un numero gran es pot apreciar a quina iteració convergeix
# La fletxa de les arestes indica la direcció de la política

num_iterations = 20
show_result(num_iterations)
```
